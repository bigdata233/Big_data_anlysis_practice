```python
#1-1
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_1_1.csv')
print(data.info())
print(data.head())
```

```python
# 1-1-1 (처리 시간 칼럼 생성)
data.loc[:, ['신고일자', '신고시각', '처리일자', '처리시각']] = data.loc[:, ['신고일자', '신고시각', '처리일자', '처리시각']].astype(str)
data['신고일시'] = pd.to_datetime(data['신고일자'] + data['신고시각'], format = '%Y%m%d%H%M%S')
data['처리일시'] = pd.to_datetime(data['처리일자'] + data['처리시각'], format = '%Y%m%d%H%M%S')
data['처리시간'] = (data['처리일시'] - data['신고일시']).dt.total_seconds()
```

```python
# 1-1-2 (공장명별 처리 시간의 평균 계산)
mean_response_time = data.groupby('공장명')['처리시간'].mean().reset_index()
print(mean_response_time)
```

```python
# 1-1-3 (데이터 정렬 후 출력)
print(mean_response_time.sort_values(by = '처리시간').iloc[2, 0])
```

```python
# 1-3
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_1_3.csv')
print(df.info())
print(df.head())
```

```python
# 1-3

# 총 판매량 칼럼 생성
df['총판매량'] = df[['제품A', '제품B', '제품C', '제품D', '제품E']].sum(axis=1)

# 분기 칼럼 생성
df1 = df.copy()
df1[['연도', '월']] = df1['기간'].str.split('_', expand=True)   # 분기를 구하기 위해 기간 칼럼에서 연도와 월을 분리한 새로운 칼럼을 생성합니다.
print(df1[['연도', '월']].head())
df1['월'] = df1['월'].str.replace('월', '').astype(int)   # 월을 숫자로 처리하기 위해 '월' 문자를 제거합니다.
df1['분기'] = pd.cut(df1['월'], bins = [0,3,6,9,12], labels = [1,2,3,4], right=True) # pd.cut()을 활용하여 분기에 해당하는 구간을 나눈 후 분기 칼럼을 생성합니다.
print(df1.head())

# 분기별 총판매량의 평균 계산
quarterly_avg = df1.groupby(['연도', '분기'])['총판매량'].mean().reset_index()

# 최댓값을 갖는 연도, 분기 찾기
print(quarterly_avg.loc[quarterly_avg['총판매량'].idxmax()])
```

```python
#2
import pandas as pd
import numpy as np
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_2_test.csv')
print(train.info())
print(test.info())
```

```python
# 2-1 (독-종분리)
test_id = test['ID']

train_X = train.drop('Target', axis = 1)
train_y = train['Target']
test_X = test.drop('Target', axis=1)
test_y = test['Target']
```

```python
# 2-2 (성능데이터 분리)
from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y = train_test_split(train_X, train_y, test_size=0.3, random_state=1)
```

```python
# 2-3 (cat-num 분리)
cat_columns = train_X.select_dtypes('object').columns.to_list()
num_columns = train_X.select_dtypes('number').columns.to_list()
```

```python
# 2-4 (라벨인코더,std스케일러 불러오기)
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
ordinalencoder = OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1)
stdscaler = StandardScaler()
```

```python
# 2-5 (라벨인코더, std스케일러 전처리)
train_X_category = ordinalencoder.fit_transform(train_X[cat_columns])
valid_X_category = ordinalencoder.transform(valid_X[cat_columns])
test_X_category = ordinalencoder.transform(test_X[cat_columns])

train_X_number = stdscaler.fit_transform(train_X[num_columns])
valid_X_number = stdscaler.transform(valid_X[num_columns])
test_X_number = stdscaler.transform(test_X[num_columns])

train_X_preprocessed = np.concatenate([train_X_number, train_X_category], axis=1)
valid_X_preprocessed = np.concatenate([valid_X_number, valid_X_category], axis=1)
test_X_preprocessed = np.concatenate([test_X_number, test_X_category], axis=1)
```

```python
# 2-6 (랜포 모델 적합)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 1)
rf.fit(train_X_preprocessed, train_y)
```

```python
# 2-7 (검증, 성능 확인)
from sklearn.metrics import f1_score
pred_val = rf.predict(valid_X_preprocessed)
print(f1_score(valid_y, pred_val, average = 'macro'))
```

```python
# 2-8 (최종 예측)
test_pred = rf.predict(test_X_preprocessed)
test_pred = pd.DataFrame(test_pred, columns = ['pred'])
```

```python
# 2-9 (파일 생성, 저장)
result = pd.concat([test_id, test_pred], axis=1)
print(result)
result.to_csv('result.csv', index=False)
```

```python
# 3-1
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_3_1.csv')
print(df.info())
```

```python
# 3-1-1
mean_before = df['before'].mean()
std_before = df['before'].std()

mean_after = df['after'].mean()
std_after = df['after'].std()

print(round(mean_before, 2), round(std_before, 2), round(mean_after, 2), round(std_after, 2))
```

```python
# 3-1-2 (대응표본 t-검정)
from scipy.stats import ttest_rel
statistic, p_value = ttest_rel(df['before'], df['after'])
print(round(statistic,2))
```

```python
# 3-1-3 (기각/채택 구하기)
p_value = p_value.round(3)
if p_value < 0.05:
  result = '기각'
else:
  result = '채택'
print(result)
```

```python
# 3-2
import pandas as pd
import numpy as np
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_3_2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/7_3_2_test.csv')
print(train.info())
print(test.info())
```

```python
# 3-2-1 (로지스틱 적합, 소득 변수의 오즈비)
import statsmodels.api as sm
import statsmodels.formula.api as smf

formula = 'purchase ~ age + income + family_members'
model = smf.logit(formula, data = train).fit()
print(model.summary())

odds = model.params['income']
odds_ratio = np.exp(odds)
print(odds_ratio)
```

```python
# 3-2-2 (Residual Deviance 구하기)
residual = -2 * model.llf
print(residual)
```

```python
# 3-2-3 (test데이터로 오분류율 구하기)

# 독립-종속 변수 설정, 상수항 추가(test데이터)
X_test = test[['age', 'income', 'family_members']]
y_test = test['purchase']
X_test = sm.add_constant(X_test)

# 예측수행
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob >= 0.5).astype(int)

# 오류율 계산
error_rate.mean = y_pred != y_test
print(round(error_rate, 3))
```
