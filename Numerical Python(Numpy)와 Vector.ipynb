```python
import numpy as np

a=np.array([1,2,3,4,5])   #숫자형 벡터
b=np.array(["apple", "banana", "orange "])    #문자형 벡터
c=np.array([True, False, True, True])   #논리형 벡터

print("Numeric Vector:", a)
print("String Vector:", b)
print("Boolean Vector:", c)
```

```python
# 숫자 8을 4번 반복
repeated_vals = np.repeat(8, 4)
print("Repeated 8 four times:", repeated_vals)
```

```python
# 배열 [1,2,4]를 2번 반복
repeated_array = np.repeat([1,2,4], 2) # 배열을 평평하게 만든 후 반복
print("Repeated each element in [1,2,4] two times:", repeated_array)

# 배열 [1,2,4]의 각 요소를 각각 1,2,3번 반복
repeated_each = np.repeat([1,2,4], [1,2,3])
print("Repeated array [1, 2, 4] two times:", repeated_each)
```

```python
# 배열 [1,2,4]를 2회 반복(벡터 전체를 반복)
repeated_whole = np.tile([1,2,4], 2)
print("벡터 전체를 두 번 반복:", repeated_whole)
```

```python
import numpy as np
a = np.array([1,2,3,4,5])

sum_a = np.sum(a)   # 합계 계산
mean_a = np.mean(a)   # 평균 계산
median_a = np.median(a)   # 중앙값 계산
std_a = np.std(a, ddof=1)   # 표준편차 계산

print(sum_a, mean_a, median_a, std_a)
```
