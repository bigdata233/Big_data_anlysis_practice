```python
#1-1
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_1.csv')

#print(df.info())
#print(df.head())

df1 = df.sort_values(by = 'lotSizeSqFt', ascending = False)
df2 = df1[0:10]
replacement = df2['lotSizeSqFt'].min()
df2['lotSizeSqFt'][0:10] = replacement

result = df1.loc[df1['yearBuilt'] >= 2000, 'lotSizeSqFt'].mean().round()
print(result)


#1-2
#print(df.info())
before_std = df['numOfBathrooms'].std()

df1 = df.copy()
df1['numOfBathrooms'] = df1['numOfBathrooms'].fillna(df['numOfBathrooms'].median())
after_std = df1['numOfBathrooms'].std()

result = abs(after_std - before_std)
print(round(result, 2))


#1-3
mu = df['MedianStudentsPerTeacher'].mean()
v_std = 1.5 * df['MedianStudentsPerTeacher'].std()
cond1 = (df['MedianStudentsPerTeacher'] < mu - v_std)
cond2 = (df['MedianStudentsPerTeacher'] > mu + v_std)
cond_last = (cond1 | cond2)
df3 = df.loc[cond_last, :]
result1 = df3['MedianStudentsPerTeacher']
result2 = len(df3['MedianStudentsPerTeacher'])
print(result1)
print(result2)
```

```python
#2
import pandas as pd
import numpy as np
train_X = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_2_trainX.csv')
train_y = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_2_trainy.csv')
test_X = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_2_testX.csv')

# 2-1 (인포)
print(train_X.info())
print(train_y.info())
print(test_X.info())

# 2-2 (검증 데이터 생성)
from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y = train_test_split(train_X, train_y, test_size=0.3, random_state=1)

# 2-3 (넘-캣 분리)
num_columns = train_X.select_dtypes('number').columns.to_list()
cat_columns = train_X.select_dtypes('object').columns.to_list()

# 2-4 (전처리)
from sklearn.preprocessing import StandardScaler
stdscaler = StandardScaler()

train_X_std = stdscaler.fit_transform(train_X[num_columns])
valid_X_std = stdscaler.transform(valid_X[num_columns])
test_X_std = stdscaler.transform(test_X[num_columns])

# 2-5 (모델 적합)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1)
rf.fit(train_X_std, train_y)

# 2-6 (검증, 성능평가)
from sklearn.metrics import f1_score
pred_val = rf.predict(valid_X_std)
print(f1_score(valid_y, pred_val, average='macro'))

# 2-7 (최종 예측)
test_pred = rf.predict(test_X_std)
result = pd.DataFrame(test_pred, columns = ['pred'])

# 2-8 (생성, 저장)
result.to_csv('result.csv', index=False)
print(pd.read_csv('result.csv').head())
```

```python
#3-1
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_3_1.csv')
#rint(df.info())
#print(df.head())
#3-1-1
mean_score = round(df['score'].mean(), 2)
std_score = round(df['score'].std(), 2)

print(mean_score, std_score)

#3-1-2
from scipy.stats import ttest_1samp
statistic, p_value = ttest_1samp(df['score'], 75)
print(round(statistic, 2))

#3-1-3
p_value = p_value.round(2)
if p_value < 0.05:
  result = '기각'
else:
  result = '채택'
print(result)

#3-2
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_3_2.csv')
print(df.info())
print(df.info())

#3-2-1
from scipy.stats import chi2_contingency
cross_tab = pd.crosstab(df['gender'], df['club_membership'])

statistic, p_value, dof1, expected = chi2_contingency(cross_tab)
print(statistic.round(2))

#3-2-2(마지막)
p_value = p_value.round(2)
if p_value < 0.05:
  result = '기각'
else:
  result = '채택'

print(result)
```
