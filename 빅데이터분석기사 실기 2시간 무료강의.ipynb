```python
from sklearn.preprocessing import StandardScaler
stdscaler = StandardScaler()
df['stdscaler'] = stdscaler.fit_transform(df[['mpg']])
df.head()
```

```python
# 확인
print(df['stdscaler'].mean(), df['stdscaler'].std())  #모표준편차기준임
```

```python
# 직접계산 (표본표준편차)
std = df['mpg'].std()
mean_mpg = df['mpg'].mean()
df['stdscaler_self'] = (df['mpg']-mean_mpg) / std
df.head()
```

```python
# 확인
print(df['stdscaler_self'].mean(), df['stdscaler_self'].std())  #표본표준편차 사용
```

```python
np_std = np.std(df['mpg'])    #기본값 ddof = 0, 모표준편차(분모의 자유도n)

# numpy와 직접계산은 !표본!
np_std1 = np.std(df['mpg'], ddof=1)   #ddof = 1 표본표준편차(분모의 자유도 n-1)
df_std = df['mpg'].std()              #std()    표본표준편차(분모의 자유도 n-1)

print(np_std)
print(np_std1)
print(df_std)
```

```python
# 날짜 다루기
df = pd.DataFrame({
    '날짜' : ['20230105','20230105','20230223','20230223','20230312','20230422', '20230511'],
    '물품' : ['A','B','A','B','A','B','A'],
    '판매수' : [5,10,15,15,20,25,40],
    '개당수익' : [500,600,500,600,600,700,600]
})
df
```

```python
print(df.info())
```

```python
# 데이터 타입 datetime 으로 변경
df['날짜'] = pd.to_datetime(df['날짜'])
df.info()
```

```python
# 년,월,일,변수(열) 추가하기
df['year'] = df['날짜'].dt.year
df['month'] = df['날짜'].dt.month
df['day'] = df['날짜'].dt.day
df
```

```python
# 날짜 구간 필터링
df[df['날짜'].between('2023-01-01', '2023-02-23')]  # 좌우 모두 포함
# (주의) 날짜와 시간이 같이 있는 데이터에 between 함수를 쓸 경우 형식이 동일하게 (날짜 + 시간) 필터링 해야함
# ex: 2023-01-05 12:30:05 => between('2023-01-05 12:00:00', '2023-01-05 12:40:00') OK
#                         => between('2023-01-05', '2023-01-05') NG
```

```python
# 날짜를 인덱스로 설정 후 loc 함수 사용
df = df.set_index('날짜') #drop=True(디폴트)
df.head(3)
```

```python
print(df.loc['2023-01-05':'2023-02-23'])    #둘다 기간 포함
print(df.loc[ (df.index >= '2023-01-05') & (df.index <= '2023-02-23')])
```

```python
# 시간 데이터 만들기
df = pd.DataFrame({
    '물품' : ['A','B','A','B','A','B','A'],
    '판매수' : [5,10,15,15,20,25,40],
    '개당수익' : [500,600,500,600,600,700,600]
})
time = pd.date_range('2023-09-24 12:25:00', '2023-09-25 14:45:30', periods = 7)
df['time'] = time
df = df[['time', '물품', '판매수', '개당수익']]
df
```

```python
#1 시간이 '변수'로 있을 경우 (between 함수)
df = df[df['time'].between('2023-04-24 12:25:00', '2023-09-24 21:11:50')]
print(df)
```

```python
#2 시간이 '인덱스'로 있을 경우(between_time, loc함수)

# 1) df.between_time(start_time ='00:00:00', end_time = '00:00:00')
# ★ 날짜와 상관없이 특정 시간대를 필터링 해야할 때
# (★ 주의 : 시간이 index에 위치해야 함)

# index 새로 지정
df1 = df.set_index('time')
# print(df1)

df1 = df1.between_time(start_time = '12:25:00', end_time = '21:11:50')
print(df1)
```

```python
# 2) loc함수 사용
print(df1.loc['2023-09-24 12:25:00':'2023-09-24 21:11:50'])
print(df1.loc[ (df1.index >= '2023-09-24 12:25:00') & (df1.index <= '2023-09-24 21:11:50')])
```
