```python
#8

# 결측치 확인
print(df.isna().sum())

# 결측치를 포함한 행 제거

df.cleaned = df.dropna()
print(df.cleaned.info())
```

```python
#9
additional_data = {
    'student_id': ['1','3','5','7','9'],
    'club': ['Art', 'Science', 'Math', 'Music', 'Drama']
}
df_additional = pd.DataFrame(additional_data)

merged_df = pd.merge(df, df_additional, on = 'student_id', how='left')
print(merged_df)
```

```python
#10
df10 = df.copy()
df_pivot_table = df10.pivot_table(index = 'gender',
                                  columns = 'student_id',
                                  values = 'average')
print(df_pivot_table)
```

```python
#11
#df11 = df.copy()
#df11['average'] = (df11['midterm']+df11['final']+df11['assignment']) / 3
#print(df11)
#df11.groupby('gender').mean(numeric_only=True)

# average 열 추가(7번과 마찬가지로 .mean(axis=1)활용)
df['average'] = df[['midterm', 'final', 'assignment']].mean(axis=1)

# melt 함수를 활용하여 데이터 프레임 변환
melted_df = pd.melt(df,
                    id_vars = ['student_id', 'name', 'gender'],
                    value_vars= ['midterm', 'final', 'assignment', 'average'],
                    var_name = 'variable',
                    value_name = 'score')

# gender, variable를 기준으로 그룹화하고 평균 점수 계산
grouped_mean = melted_df.groupby(['gender', 'variable'])['score'].mean().reset_index()
print(grouped_mean)
```
