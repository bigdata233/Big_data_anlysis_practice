```python
import numpy as np
a = np.array([1,2,3,4,5])
b = a + 5
print(b)
```

```python
a = np.array([12,21,35,48,5])
c = a[np.arange(1,6) % 2 == 1]
print(c)
```

```python
a = np.array([1,22,93,64,54])
c = np.max(a)
print(c)
```

```python
a = np.array([1,2,3,2,4,5,4,6])
d = np.unique(a)
print(d)
```

```python
a = np.array([21,31,58])
b = np.array([24,44,67])
c = np.empty(a.size + b.size, dtype =a.dtype)
c[0::2] = a
c[1::2] = b
print(c)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9])

c = a[:-1] + b
print(c)
```
