```python
# 문제1

import pandas as pd
import numpy as np

# 예제 데이터 생성
np.random.seed(42)
n_samples = 210
X = np.random.randn(n_samples, 4)
y = (X[:,0]+X[:,1]*0.5 + np.random.randn(n_samples)*0.5>0).astype(int)
df = pd.DataFrame(X, columns = ['weight', 'height', 'age', 'income'])
df['gender'] = y

# 데이터 확인
print(df.head())
```

```python
#1

import statsmodels.formula.api as smf
import statsmodels.api as sm

formula = 'gender ~ weight'
model = smf.logit(formula, data=df).fit()
print(model.summary())
odds = model.params['weight']
odds_ratio = np.exp(odds)
print(odds_ratio)
```

```python
#2

formula = 'gender ~ weight + height + age + income'
model = smf.logit(formula, data=df).fit()
residual_deviance = -2 * model.llf
print(residual_deviance)
```

```python
#3

from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size = 90, random_state=42)

train_X = sm.add_constant(df_train[['weight']])
train_y = df_train['gender']
test_X = sm.add_constant(df_test[['weight']])
test_y = df_test['gender']
print(train_X.shape)
print(test_X.shape)

model = sm.Logit(train_y, train_X).fit()

from sklearn.metrics import accuracy_score

pred_val = model.predict(test_X) > 0.5
error_rate = 1 - accuracy_score(test_y, pred_val)
print(error_rate)
```
