```python
#1-1
import pandas as pd
import numpy as np
dat = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_1_1.csv')
print(dat.head())
pd.set_option('display.max_columns', None) # 하나마나?
```

```python
#1-1-1 (.groupby를 통해 대륙별 평균 맥주소비량을 구한 후 가장 맥주소비량이 많은 대륙을 구합니다.)
result = dat.groupby('대륙')['맥주소비량'].mean().idxmax()
print(result)
```

```python
#1-1-2 (맥주소비량이 가장 많은 대륙 데이터를 필터링하고 5번째로 맥주소비량이 많은 나라를 계산합니다.)
sub_dat = dat.loc[dat['대륙'] == result]
result2 = sub_dat.sort_values(by = '맥주소비량', ascending = False).iloc[4,1]
print(result2)
```

```python
#1-1-3 (5번째로 맥주소비량이 많은 나라의 평균 맥주 소비량 구하기)
result3 = dat.loc[dat['국가'] == result2, '맥주소비량'].mean().round()
print(result3)
```

```python
#1-2
dat = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_1_2.csv')
print(dat.head())
```

```python
#1-2-1 (관광객비율이 두 번째로 높은 나라의 관광 수를 구하기)

# 문제에 정의된 기준에 맞춰서 합계 칼럼을 생성합니다. axis=1은 행방향 합계, axis=0 열방향 합계

dat['합계'] = dat.loc[:, ['관광', '사무', '공무', '유학', '기타']].sum(axis=1)

# 문제에 정의된 기준에 맞춰서 관광객비율 칼럼을 생성합니다.

dat['관광객비율'] = np.round(dat['관광'] / dat['합계'], 3)

# 관광객비율이 두 번째로 높은 나라의 관광 수를 계산합니다.(iloc 사용)

result1 = dat.sort_values(by = '관광객비율', ascending = False).iloc[1,1]
print(result1)
```

```python
# 1-2-2 (관광 수가 두 번째로 높은 나라를 찾고 그 나라의 공무 수 평균을 계산합니다.)

second_country = dat.sort_values(by = '관광', ascending = False).iloc[1, 0]
result2 = dat.loc[dat['국가'] == second_country, '공무'].mean().round()
print(result2)
```

```python
# 1-2-3  (이전에 구한 관광 수와 공무 수 합계 구하기)
result3 = result1 + result2
print(result3)
```

```python
# 2
import pandas as pd
import numpy as np
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_2_test.csv')
```

```python
# 2-1 (인포)
print(train.info())
print(test.info())
```

```python
# 2-2 (id칼럼 분해, 독-종 분해)

test_id = test['ID']
train_X = train.drop('count', axis=1)
train_y = train['count']
test_X = test.drop('count', axis=1)
test_y = test['count']
```

```python
# 2-3 (검증, 트레인-테스트 분리)
from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y = train_test_split(train_X, train_y, test_size= 0.3, random_state = 1)
```

```python
# 2-4 (원-핫 전처리)
cat_columns = train_X.select_dtypes('object').columns.to_list()
num_columns = test_X.select_dtypes('number').columns.to_list()
```

```python
# 2-5 (원-핫 불러오기)
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(sparse_output = False, handle_unknown= 'ignore')
```

```python
# 2-6 (원-핫 수행하기)
train_X_preprocessed = onehotencoder.fit_transform(train_X[cat_columns])
valid_X_preprocessed = onehotencoder.transform(valid_X[cat_columns])
test_X_preprocessed = onehotencoder.transform(test_X[cat_columns])
```

```python
# 2-7 (랜덤포레스트 적합하기)
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=1)
rf.fit(train_X_preprocessed, train_y)
```

```python
# 2-8 (검증, 모델 성능 확인하기)
from sklearn.metrics import mean_absolute_error
pred_val = rf.predict(valid_X_preprocessed)
print(mean_absolute_error(valid_y, pred_val))
```

```python
# 2-9 (최종 예측 수행하기)
test_pred = rf.predict(test_X_preprocessed)
test_pred = pd.DataFrame(test_pred, columns = ['pred'])
```

```python
# 2-10 (생성, 저장하기)
result = pd.concat([test_id, test_pred], axis=1)
result.to_csv('result.csv', index=False)
print(result)
```

```python
# 3-1
import pandas as pd
import numpy as np
dat = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_3_1.csv')
print(dat.head())
print(dat.info())
```

```python
# 3-1-1 (도입 전, 도입 후의 평균과 표준편차 구하기) -> round활용 조심
mean_before = dat['before'].mean()
std_before = dat['before'].std()
mean_after = dat['after'].mean()
std_after = dat['after'].std()

print(round(mean_before, 2))
print(round(std_before, 2))
print(round(mean_after, 2))
print(round(std_after, 2))
```

```python
# 3-1-2 (부호 순위 검정 실시, 검정통계량 구하기 **)
from scipy.stats import wilcoxon
statistic, p_value = wilcoxon(dat['before'], dat['after'])
print(statistic)
```

```python
# 3-1-3 (귀무가설의 기각 / 채택여부 파악)
p_value = p_value.round(3)
# 결과 해석
if p_value < 0.05:
  result = '기각'
else:
  result = '채택'

print(result)
```

```python
# 3-2
import pandas as pd
import numpy as np
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_3_2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/8_3_2_test.csv')
print(train.head())
print(test.head())
```

```python
# 3-2-1 (다중회귀모형 적합 후, 회귀계수 최대인 변수 구하기(상수항 포함))
import statsmodels.api as sm
from statsmodels.formula.api import ols
formula = 'productivity ~ hours + age + experience'
multi_ols = ols(formula, data = train).fit()
print(multi_ols.summary())

coef = multi_ols.params[1:]
print(coef.idxmax())
```

```python
# 3-2-2 (유의수준 5% 하에서 p-value를 확인하고, 유의미한 변수의 개수를 계산합니다.)

p_values = multi_ols.pvalues[1:].round(4)
significant_p_values = p_values[p_values < 0.05]
num_significant_p_values = len(significant_p_values)
print(significant_p_values, num_significant_p_values)
```

```python
# 3-2-3 (테스트 데이터로 모델의 성능을 평가하시오)

from sklearn.metrics import r2_score

# 독립 변수와 종속 변수 설정하기
X_test = test[['hours','age','experience']]
y_test = test['productivity']
X_test = sm.add_constant(X_test)

y_pred = multi_ols.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(round(r2, 3))
```
